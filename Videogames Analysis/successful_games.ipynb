# Сборный проект 1. Исследование игр.

**Задача:**  

Выявить определяющие успешность
игры закономерности. Это позволит сделать ставку на потенциально популярный
продукт и спланировать рекламные кампании.

**Описание данных:**
- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). 
Данные за 2016 год могут быть неполными.

## Изучение данных из файла

```python
import pandas as pd
import seaborn as sns
import warnings
import numpy as np
from scipy import stats as st
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
%config InlineBackend.figure_format='retina'
%matplotlib inline
```

```python
try: 
    df = pd.read_csv('datasets/games.csv')
except:
    df = pd.read_csv('/datasets/games.csv')
```

В данных мы видим проблемы:
1. некорректно названные столбцы (uppercase),
2. некорректные типы данных 
- год выпуска должен быть целочисленным или datetime,
- critic_score - integer,
- user_score - float.
3. Пропущенные значения

```python
# сразу приведем названия столбцов к нижнему регистру
df.columns = df.columns.str.lower()
```

### Дубликаты

Проверим данные на явные дубликаты.

Явных дубликатов нет.

Проверим данные на неявные дубли. Предположительно, они могут быть в столбцах platform и genre, т.к. они текстовые.

```python
df['platform'].value_counts().sort_values(ascending=False)
```

```python
df['genre'].value_counts().sort_values(ascending=False)
```

Неявных дублей тоже нет. 

### Пропуски

```python
df.isna().mean()
```

В данных достаточно много пропусков:
- 51% в critic_score
- 40% в user_score и rating
- 1.6% в year_of_release
- 0.01% в genre и name

1. Для дальнейшего исследования нам будут нужны жанры и год выпуска.  Так как их некорректно заполнить ни нулем, ни средним, а процент пропущенных данных небольшой, логично будет от них избавиться.

2. Значения rating, critic_score и user_score - мы не можем ничем заполнить, т.к. пропусков слишком много и заполнение по средним значениям исказит статистику.

```python
df = df.dropna(subset=['year_of_release', 'genre'])
```

```python
# заменим тип данных на корректный
df['year_of_release'] = df['year_of_release'].astype('int64')
```

```python
df.isna().mean()
```

**Посмотрим на пропуски в user_score**

```python
df['user_score'].value_counts()
```

В столбце user_score есть значение tbd - To Be Determined. Скорее всего аббревиатура служит для отметки неясностей или пропусков, которые надо заполнить.
Необходимо преобразовать его в nan.

```python
df['user_score'] = df['user_score'].replace(['tbd'], np.nan)
```

```python
# приведем значения user_score к вещественному типу
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
```

### Суммарные продажи в регионах

Рассчитаем и добавим в таблицу суммарное значение продаж по всем регионам для каждой игры. 

```python
df['total_sales'] = df[['na_sales','eu_sales','jp_sales','other_sales']].sum(axis=1)
```

```python
df.head()
```

## EDA

### Выпуск игр по годам

```python
plt.figure(figsize = (15,5))
sns.histplot(data = df, x='year_of_release')
plt.show()
```

### Продажи по платформам

Рассмотрим ТОП 10 платформ по популярности и изучим, как изменялись продажи  со временем.

```python
top_platf = df_new.groupby('platform')['total_sales'].sum().sort_values(ascending=False)[:10]
top_platf
```

Больше всего продаж игр было для платформ PS2, X360, PS3

Посмотрим на распределение продаж по ТОП 10 платформам со временем.

```python
top_platforms_all_years = (
    df_new
    .query('platform in @top_platf.index')
    .pivot_table(index=['platform', 'year_of_release'], values = 'total_sales', aggfunc = 'sum')
    .reset_index()
)
```

```python
plt.figure(figsize = (15,5))
sns.lineplot(data = top_platforms_all_years, x='year_of_release', y = 'total_sales',hue = 'platform')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.show()
```

Можно наблюдать, как каждая платформа живет в среднем от 5 до 10 лет. 
Потом устаревает и либо выходит новая версия, либо выходит новая версия у конкурентов.
Определим этот актуальный период более точно.

### Определение актуального периода

Определим минимальный порог продаж в 1 млн, посчитаем сколько в среднем лет доход платформы превышает порог и найдем актуальный период.

```python
#вспомогательная сводная
top_platforms = df_new.pivot_table(index=['platform', 'year_of_release'], values = 'total_sales', aggfunc = 'sum').reset_index()
```

```python
# выбираем строки с годами, где продаж было больше 1 млн
# еще раз сводим, считаем количество таких годов
# усредняем по всем платформам
# делим на 2 - половину жизненного цикла платформы растут, а вторую половину - падают
years = (
    top_platforms.query('total_sales >1')
    .pivot_table(index='platform', values = 'total_sales', aggfunc = 'count')
    ['total_sales'].mean()/2
)
years
```

Усредненно возьмем актуальный период в 4 года, т.е. актуальные данные для 2017 года - после 2012. 

```python
df_actual = df.query('year_of_release > 2012')
```

## Портрет пользователя каждого региона

### Топ 5 популярных платформ по регионам

```python
regions = (
    df_actual
    .pivot_table(index='platform', values = ['eu_sales','jp_sales', 'na_sales','other_sales', 'total_sales'], aggfunc='sum')
    .sort_values(by='total_sales', ascending=False)
    .reset_index()
    [:5]
)
regions
```

```python
regions.plot(x='platform', y= ['eu_sales','jp_sales', 'na_sales','other_sales'], kind="bar", figsize = (15,5), grid=True)
```

PS4 и PS3 наиболее популярны в Европе, а XOne и X360 - в Северной Америке. 
В Японии наиболее популярна 3DS

Посмотрим на распределение по регионам.

```python
labels = regions['platform']
fig = plt.figure(figsize=(15,5))
plt.subplot (1, 4, 1)
plt.pie(x=regions['eu_sales'], autopct='%.1f%%', labels=labels)
plt.title("Europe Sales")
plt.subplot (1, 4, 2)
plt.pie(x=regions['jp_sales'], autopct='%.1f%%', labels=labels)
plt.title("Japan Sales")
plt.subplot (1, 4, 3)
plt.pie(x=regions['na_sales'], autopct='%.1f%%', labels=labels)
plt.title("North America Sales")
plt.subplot (1, 4, 4)
plt.pie(x=regions['other_sales'], autopct='%.1f%%', labels=labels)
plt.title("Other Sales")
fig.tight_layout()
plt.show()
```

- В Европе наиболее популярна PS4
- В Японии больше половины рынка занимает 3DS
- В Америке Playstation и XBOX делят рынок примерно пополам
- Во всех остальных странах однозначный лидер PlayStation - почти 70% рынка.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Проверим гипотезы:

Н0: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые  
Н1: Средние пользовательские рейтинги платформ Xbox One и PC различаются

```python
xbox = df_actual[df_actual['platform'] == 'XOne']['user_score'].dropna()
pc = df_actual[df_actual['platform'] == 'PC']['user_score'].dropna()
```

Проверим распределение при помощи критерия Шапиро-Уилка

Н0: Данные рейтинга XBox One и PC распределены нормально 
Н1: Данные рейтинга XBox One и PC не распределены нормально

```python
alpha = .05 #5% стат значимости будет достаточно для данной тематики
results = st.shapiro(xbox)

print('p-значение: ', results.pvalue)

if results.pvalue > alpha:
    print('Выборка похожа на нормальное распределение (не получилось отвергнуть нулевую гипотезу)')
else:
    print('Выборка не похожа на нормальное распределение (отвергаем нулевую гипотезу)')
```

```python
alpha = .05 #5% стат значимости будет достаточно для данной тематики
result = st.shapiro(pc)

print('p-значение: ', results.pvalue)

if results.pvalue > alpha:
    print('Выборка похожа на нормальное распределение (не получилось отвергнуть нулевую гипотезу)')
else:
    print('Выборка не похожа на нормальное распределение (отвергаем нулевую гипотезу)')
```

Мы обнаружили, что распределение выборок можно считать нормальным.  
Значит, при проверке гипотез мы можем использовать критерии Стьюдента.

```python
alpha = .05 #5% стат значимости будет достаточно для данной тематики
results = st.ttest_ind(
    xbox, 
    pc)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")

print ('Среднее выборки xbox', xbox.mean())
print ('Среднее выборки pc', pc.mean())
```

Мы не можем с уверенностью сказать, что средние пользовательские рейтинги платформ Xbox One и PC различаются
